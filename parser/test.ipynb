```python
import dut_parser

pdf_path = "../specifications/EECS-2011-62.pdf"
dut_text = dut_parser.load_dut_spec(pdf_path)

parsed_entities = dut_parser.parse_dut_text(dut_text)

# length of the text
print("No. of characters ", len(dut_text))
print("No. of entities ", len(parsed_entities))

for entity, label in parsed_entities:
    print(f"Entity: {entity}, Label: {label}")
```

Custom tokens extracted

```python
custom_entities = [(entity, label) for entity, label in parsed_entities if label in dut_parser.ENTITY_LABELS]

print("Custom Entities Extracted Successfully:")
print(f"Total custom entities: {len(custom_entities)}")
for entity, label in custom_entities:
    print(f"Entity: {entity}, Label: {label}")
```
